```python
import uproot3 as uproot
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_HepMC_DH='D:\\ARPE\\Reco\\Tracksummary\\Tel_x\\HepMC\\DH_0.8128\\Track\\'
filename=path_HepMC_DH+'trackstates_fitter.root'
```

```python

# Open the ROOT file
file = uproot.open(filename)

# Access the TTree
tree = file["trackstates"]

# Read the branches into numpy arrays
event_nr = tree["event_nr"].array()
multiTraj_nr = tree["multiTraj_nr"].array()
subTraj_nr = tree["subTraj_nr"].array()
t_x = tree["t_x"].array()
t_y = tree["t_y"].array()
t_z = tree["t_z"].array()
t_r = tree["t_r"].array()
t_dx = tree["t_dx"].array()
t_dy = tree["t_dy"].array()
t_dz = tree["t_dz"].array()
t_eLOC0 = tree["t_eLOC0"].array()
t_eLOC1 = tree["t_eLOC1"].array()
t_ePHI = tree["t_ePHI"].array()
t_eTHETA = tree["t_eTHETA"].array()
t_eQOP = tree["t_eQOP"].array()
t_eT = tree["t_eT"].array()
nStates = tree["nStates"].array()
nMeasurements = tree["nMeasurements"].array()
volume_id = tree["volume_id"].array()
layer_id = tree["layer_id"].array()
module_id = tree["module_id"].array()
pathLength = tree["pathLength"].array()
l_x_hit = tree["l_x_hit"].array()
l_y_hit = tree["l_y_hit"].array()
g_x_hit = tree["g_x_hit"].array()
g_y_hit = tree["g_y_hit"].array()
g_z_hit = tree["g_z_hit"].array()
res_x_hit = tree["res_x_hit"].array()
res_y_hit = tree["res_y_hit"].array()
err_x_hit = tree["err_x_hit"].array()
err_y_hit = tree["err_y_hit"].array()
pull_x_hit = tree["pull_x_hit"].array()
pull_y_hit = tree["pull_y_hit"].array()
dim_hit = tree["dim_hit"].array()

g_x_smt=tree["g_x_smt"].array()
g_y_smt=tree["g_y_smt"].array()
g_z_smt=tree["g_z_smt"].array()
chi2=tree["chi2"].array()


# Flatten variables into 1D numpy arrays
event_nr_1d = event_nr.flatten()
multiTraj_nr_1d = multiTraj_nr.flatten()
subTraj_nr_1d = subTraj_nr.flatten()
t_x_1d = t_x.flatten()
t_y_1d = t_y.flatten()
t_z_1d = t_z.flatten()
t_r_1d = t_r.flatten()
t_dx_1d = t_dx.flatten()
t_dy_1d = t_dy.flatten()
t_dz_1d = t_dz.flatten()
t_eLOC0_1d = t_eLOC0.flatten()
t_eLOC1_1d = t_eLOC1.flatten()
t_ePHI_1d = t_ePHI.flatten()
t_eTHETA_1d = t_eTHETA.flatten()
t_eQOP_1d = t_eQOP.flatten()
t_eT_1d = t_eT.flatten()
nStates_1d = nStates.flatten()
nMeasurements_1d = nMeasurements.flatten()
volume_id_1d = volume_id.flatten()
layer_id_1d = layer_id.flatten()
module_id_1d = module_id.flatten()
pathLength_1d = pathLength.flatten()
l_x_hit_1d = l_x_hit.flatten()
l_y_hit_1d = l_y_hit.flatten()
g_x_hit_1d = g_x_hit.flatten()
g_y_hit_1d = g_y_hit.flatten()
g_z_hit_1d = g_z_hit.flatten()
res_x_hit_1d = res_x_hit.flatten()
res_y_hit_1d = res_y_hit.flatten()
err_x_hit_1d = err_x_hit.flatten()
err_y_hit_1d = err_y_hit.flatten()
pull_x_hit_1d = pull_x_hit.flatten()
pull_y_hit_1d = pull_y_hit.flatten()
dim_hit_1d = dim_hit.flatten()

g_x_smt_1d = g_x_smt.flatten()
g_y_smt_1d = g_y_smt.flatten()
g_z_smt_1d = g_z_smt.flatten()
chi2_1d = chi2.flatten()


```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Flatten the subplots array for easier indexing
axs = axs.flatten()

# Plot histogram of res_x_hit_1d
axs[0].hist(res_x_hit_1d, bins=50)
axs[0].set_xlabel('res_x_hit')
axs[0].set_ylabel('Frequency')

# Plot histogram of res_y_hit_1d
axs[1].hist(res_y_hit_1d, bins=50)
axs[1].set_xlabel('res_y_hit')
axs[1].set_ylabel('Frequency')

# Plot g_x_hit[0] vs g_y_hit[0]
axs[2].scatter(g_x_hit[0], g_y_hit[0], s=5)
axs[2].set_xlabel('g_x_hit[0]')
axs[2].set_ylabel('g_y_hit[0]')

# Plot g_x_hit[0] vs g_z_hit[0]
axs[3].scatter(g_x_hit[0], g_z_hit[0], s=5)
axs[3].set_xlabel('g_x_hit[0]')
axs[3].set_ylabel('g_z_hit[0]')

# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Show the plots
plt.show()
```

```python
color_hist = 'blue'
color_scatter1 = 'red'
color_scatter2 = 'green'

# Create subplots with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
colors = plt.cm.viridis(np.linspace(0, 1, 4000))
# Flatten the subplots array for easier indexing
axs = axs.flatten()

# Plot histogram of res_x_hit_1d
axs[0].hist(res_x_hit_1d, bins=100, color=color_hist)
axs[0].set_xlabel('res_x_hit')
axs[0].set_ylabel('Frequency')
axs[0].set_xlim(-200, 200)# Set x-axis limits

for k in range(0,3000):
    axs[1].scatter(g_y_hit[k], g_z_hit[k], s=5, color=colors[k])
    axs[1].scatter(g_y_smt[k], g_z_smt[k], s=5, color=colors[k+20])
axs[1].set_xlabel('g_y_hit')
axs[1].set_ylabel('g_z_hit')
axs[1].set_xlim(-2000, 2000)# Set x-axis limits
axs[1].set_ylim(-900, 900)# Set x-axis limits

# Plot g_x_hit[0] vs g_y_hit[0]
for k in range(0,600):
    axs[2].scatter(g_x_hit[k], g_y_hit[k], s=5, color=colors[k])
    axs[2].scatter(g_x_smt[k], g_y_smt[k], s=5, color=colors[k+20])
axs[2].set_xlabel('g_x_hit')
axs[2].set_ylabel('g_y_hit')
axs[2].set_xlim(0, 23000)  # Set x-axis limits
# axs[2].set_ylim(-1600, 1600)  # Set y-axis limits

# Plot g_x_hit[0] vs g_z_hit[0]
for k in range(0,600):
    axs[3].scatter(g_x_hit[k], g_z_hit[k], s=5, color=color_scatter2)
axs[3].set_xlabel('g_x_hit')
axs[3].set_ylabel('g_z_hit')
axs[3].set_xlim(0, 23000)  # Set x-axis limits
axs[3].set_ylim(-600, 600)  # Set y-axis limits

# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Show the plots
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

n=30
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = np.random.rand(n)
cmap = plt.cm.get_cmap('viridis')

fig = plt.figure(figsize=(15, 15))


# Loop through the data points and plot them in 3D
for k in range(n):
    ax.scatter(g_x_hit[k], g_y_hit[k], g_z_hit[k], s=5, c=cmap(colors[k]), label=f'Data {k+1}')

ax.set_xlabel('g_x_hit')
ax.set_ylabel('g_y_hit')
ax.set_zlabel('g_z_hit')
ax.set_xlim(0, 21000)
ax.set_ylim(-1600, 1600)
ax.set_zlim(-600, 600)
ax.legend()

plt.show()
```
